### W5A18 
VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def doi_xung(n):
    so=n
    moi=0
    while so>0:
        doi=so%10
        moi=moi*10 + doi
        so//=10
    if n==moi:
        return True
    return False
```

### W5A20 
Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
```

Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def are_collinear(x1,y1,x2,y2,x3,y3):
    xAB=x2-x1
    yAB=y2-y3
    xAC=x3-x1
    yAC=y3-y1
    if xAB/xAC==yAB/yAC:
        return True
    return False
```
