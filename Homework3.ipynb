### Bài 1:
In các số dương lẻ bé hơn 100

```python
for i in range (0,101):
    if i%2==1:
        print(i)
```

### Bài 2:
In các số dương chẵn bé hơn 100

```python
for i in range (0,101):
    if i%3==0:
        print(i)
```

### Bài 3:
Nhập và số nguyên dương a, in ra bảng cửu chương của a

```python
a=int(input())
for i in range (1,11):
    print(f"{a} x {i} = {a*i}")
```
