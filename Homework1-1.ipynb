### Bài 2: 
 Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
a,b=map(int,input().split())
r=a/2
s=a*b-3.14*(r**2)
print(f"{s:.2f}")
```

### Bài 3: 
Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma)

```python
a=input()
ma=ord(a)
if 65<=ma<=90:
    thuong=ma+32
    print(chr(thuong))
elif 97<=ma<=122:
    hoa=ma-32
    print(chr(hoa))
```

### Bài 4: 
Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} không phải là kí tự alphabet.

```python
c=input()
if len(c)==1 and (65<=ord(c)<=90 or 97<=ord(c)<=122):
    print (f"{c} là kí tự alphabet")
else:
    print (f"{c} không là kí tự alphabet")

```

### Bài 6. 
Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
import math
a,b,c=map(int,input().split())
if a+b>c and a+c>b and b+c>a:
    p=(a+b+c)/2
    s=math.sqrt(p*(p-a)*(p-b)*(p-c))
    print(f"{s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh

```python
ten=input("Ten chu ho: ")
thang_trc=int(input("Chi so thang truoc: "))
thang_nay=int(input("Chi so thang nay: "))
tieu_thu=thang_nay-thang_trc
vat=108/100
print(f"Ho va ten: {ten}")
if tieu_thu<=50:
    tien=tieu_thu*1984*vat
    print(f"Tien phai tra la: {tien:.0f}")
elif 50<tieu_thu<=100:
    tien=50*1984*vat+(tieu_thu-50)*2050*vat
    print(f"Tien phai tra la: {tien:.0f}")
elif 100<tieu_thu<=200:
    tien=50*1984*vat+50*2050*vat+(tieu_thu-100)*2380*vat
    print(f"Tien phai tra la: {tien:.0f}")
elif 200<tieu_thu<=300:
    tien=50*1984*vat+50*2050*vat+100*2380*vat+(tieu_thu-200)*2998*vat
    print(f"Tien phai tra la: {tien:.0f}")
elif 300<tieu_thu<=400:
    tien=50*1984*vat+50*2050*vat+100*2380*vat+100*2998*vat+(tieu_thu-300)*3350*vat
    print(f"Tien phai tra la: {tien:.0f}")
else:
    tien=50*1984*vat+50*2050*vat+100*2380*vat+100*2998*vat+100*3350*vat+(tieu_thu-400)*3460*vat
    print(f"Tien phai tra la: {tien:.0f}")
```
