### W3A1
Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
a=int(input())
box=0
if 100<=a<=999:
    while a>0:
        box=box*10+a%10
        a//=10
    print(box)
```

### W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
hay không.  
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
n=int(input())
if n&(n-1)==0:
    print("YES")
else:    print("NO")
```

### W3A5 
Viêt chương trình chia một số m cho n, kết quả thu được làm tròn lên

```python
m,n=map(int,input().split())
if m%n==0:
    print(m/n)
else:  
    print(m//n+1)
```

### W3A6
[Even and Odd]  
Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ.
Nếu x là số chẵn, in ra "Even". Nếu x là số lẻ, in ra "Odd".

```python
x=int(input())
if x>=0:
    if x%2==0:
        print("EVEN")
    else: print("ODD")
```

### W3A7
[Check negative numbers]  
Nhập vào hai số nguyên a, b. Nếu cả hai số a, b đều âm thì in ra "Yes",
ngược lại in ra "No".

```python
a,b=map(int,input().split())
if a<0 and b<0:
    print('YES')
else: print('NO')
```

### W3A8
[Longer string]. 
Nhập vào hai chuỗi a, b gồm các chữ cái thường. In ra "True" nếu chuỗi a
dài hơn chuỗi b. Ngược lại in ra "False".

```python
while True:
    a,b=map(int,input().split())
    if 0<a<=b:
        break
print(a,b)
```
