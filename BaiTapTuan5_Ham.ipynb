### W5A1
Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> int:

```python
def Max_Of_Two( a, b) ->int:
    Max=a
    if b>Max:
        Max=b
    return Max
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

Viết hàm hoán đổi hai số nguyên. def swap(a, b) -> Tuple[int, int]:

```python
def swap(a, b) -> tuple[int, int]:
    return b, a
a, b = map(int, input().split())
a, b = swap(a, b)
print(a, b)
```

### W5A3 
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def snt(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
n = int(input())
if snt(n):
    print("True")
else:    print("False")
```

### W5A4
Viết hàm kiểm tra xem 1 số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên


```python
def hoan_hao(n):
    tong=0
    for i in range(1,n):
        if n%i==0:
            tong+=i
    return tong==n
n=int(input())
if hoan_hao(n):
    print("True")
else:    print("False")
```

### W5A5
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

```python
def tim_k(so,k):
    for i in range(len(so)):
        if so[i]==k:
            return i
    return -1

so=list(map(int,input().split()))
k=int(input())
print(tim_k(so,k))
```

### W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n.

```python
def giai_thua(n):
    for i in range(1,n):
        n*=i
    return n
n=int(input())
print(giai_thua(n))
```

Viết hàm giả lập một máy tính bỏ túi.
Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)  
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:  
Phép trừ \( - \)  
Phép cộng \( + \)  
Phép chia \( / \)  
Phép nhân \( * \)  
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
Đầu ra: Kết quả phép toán.


```python
def may_tinh(num1: float,num2: float, phep_tinh: str):
    if phep_tinh == '+':
        return num1 + num2
    elif phep_tinh == '-':
        return num1 - num2
    elif phep_tinh == '*':
        return num1 * num2
    elif phep_tinh == '/':
        if num2 != 0:
            return num1 / num2
        else:
            raise False
    else:
        return None
num1=float(input())
phep_tinh=input()
num2=float(input())
print(may_tinh(num1, num2, phep_tinh))
```

### W5A9 
Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def tong(n):
    du=0
    while n>0:
        du+=n%10
        n//=10
    return du
n=int(input())
print(tong(n))
```
