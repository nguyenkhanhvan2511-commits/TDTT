### W5A1
Viết hàm truyền vào hai số nguyên a, b. Trả về số lớn hơn trong hai số. def Max_Of_Two( a, b) -> int:

```python
def Max_Of_Two( a, b) ->int:
    Max=a
    if b>Max:
        Max=b
    return Max
a,b=map(int,input().split())
print(Max_Of_Two(a,b))
```

### W5A3 
Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def snt(n):
    if n<2:
        return False
    for i in range (2,int(n**0.5)+1):
        n%i==0
        return False
    return True
n=int(input())
if snt(n):
    print('True')
else:
    print('False')
```

### W5A4
Viết hàm kiểm tra xem 1 số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên


```python
def hoan_hao(n):
    tong=0
    for i in range(1,n):
        if n%i==0:
            tong+=i
    if tong==n:
        return True
    return False
n=int(input())
if hoan_hao(n): print('True')
else: print('False')
```

### W5A5
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử
đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1

```python
def tim_k(n,k):
    for i in range (0,len(n)+1):
        if n[i]==k:
            return i
    return -1
n=list(map(int,input().split()))
k=int(input())
print(tim_k(n,k))
```

### W5A6
Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n.

```python
def giai_thua(n):
    tich=1
    for i in range(1,n+1):
        tich*=i
    return tich
n=int(input())
print(giai_thua(n))
```

\)  
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:  
Phép trừ \( - \)  
Phép cộng \( + \)  
Phép chia \( / \)  
Phép nhân \( * \)  
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phép toán cần thực hiện (trong số các phép toán \(+, -, *, /\)).
Đầu ra: Kết quả phép toán.


```python
def may_tinh(num1,num2,operat):
    if operat=='+':
        return num1+num2
    elif operat=='-':
        return num1-num2
    elif operat=='/':
        return num1/num2
    elif operat=='*':
        return num1*num2
num1=int(input())
operat=input()
num2=int(input())
print(f'{may_tinh(num1,num2,operat):.2f}')
```

### W5A9 
Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def tong(n):
    tong=0
    while n>0:
        cuoi=n%10
        n//=10
        tong+=cuoi
    return tong
n=int(input())
print(tong(n))
```

### W5A12
Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.

```python
def divide_with_remainder(dividend, divisor):
    q=dividend//divisor
    r=dividend%divisor
    return q,r
```

### W5A14 
Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op.

```python
def compute(a, b, op):
    if op=='+':
        return a+b
    elif op=='-':
        return a-b
    elif op=='*':
        return a*b
    elif op=='/':
        return a/b
```

### W5A15 
Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông.

```python
def is_right_triangle(a, b, c):
    if a+b>c and a+c>b and b+c>a:
        if a**2==b**2+c**2 or b**2==c**2+a**2 or c**2==b**2+a**2:
            return True
    return False
```
